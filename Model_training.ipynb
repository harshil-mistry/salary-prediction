```python
objects = []
for col in df.columns:
    if df[col].dtypes not in ['int64', 'float64']:
        objects.append(col)
```

```python
objects
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoders = {}
for col in objects:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
encoders['Gender'].inverse_transform([0, 1])
```

```python
x = df.drop(columns=['Salary'])
y = df['Salary']
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
```

```python
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error (MAE): ${mae:,.2f}")
print(f"Mean Squared Error (MSE): ${mse:,.2f}")
print(f"Root Mean Squared Error (RMSE): ${rmse:,.2f}")
print(f"R-squared (R2): {r2:.4f}")
```

```python
sample_df = pd.DataFrame({'Actual': y_test.head(10), 'Predicted': y_pred[:10].round(2)})
print(sample_df)
```

```python
import joblib

model_dir = 'model_assets'
os.makedirs(model_dir, exist_ok=True) 

model_path = os.path.join(model_dir, 'random_forest_model.joblib')
joblib.dump(rf_model, model_path)
print(f"Random Forest Model exported to: {model_path}")

encoders_path = os.path.join(model_dir, 'label_encoders_dict.joblib')
joblib.dump(encoders, encoders_path)
print(f"Label Encoders dictionary exported to: {encoders_path}")

print("\nExport complete. These files are ready for deployment.")
```
